```python
import networkx as nx

import tdshub as hub
import smash

s = hub.gridlabd.feeders.load(hub.cl.gridlabd.feeders.ieee_node4())
m = hub.gridlabd.feeders.load(hub.cl.gridlabd.feeders.ieee_node13())
l = hub.gridlabd.feeders.load(hub.cl.gridlabd.feeders.gc_12_47_1_1_IGMS())
```

```python
def compute_density(pos):
    xvalues = [ y[0] for x,y in pos.items()]
    yvalues = [ y[1] for x,y in pos.items()]
    area = (max(xvalues)-min(xvalues))*(max(yvalues)-min(yvalues))
    return area/len(pos)

sp = nx.graphviz_layout(s.graph, prog="neato")
mp = nx.graphviz_layout(m.graph, prog="neato")
lp = nx.graphviz_layout(l.graph, prog="neato")

compute_density(sp), compute_density(mp), compute_density(lp)
```

```python
smash.graph(s.graph, size=8, standalone=False)
```

```python
smash.graph(m.graph, size=4, font_size=3)
```
