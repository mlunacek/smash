#Smash examples for the notebook

```python
import smash
import pandas as pd
import os
cars = pd.read_csv(os.path.join('data','cars.csv'))
```

##Scatter plot

Includes tooltip, but it's buggy.

```python
subset = cars[['economy (mpg)','weight (lb)','cylinders']]
smash.scatter(data=subset, height=400, width=600).show()
```

##Parallel Coordinates

Includes brushing.

```python
smash.parallel(data=cars).show()
```
